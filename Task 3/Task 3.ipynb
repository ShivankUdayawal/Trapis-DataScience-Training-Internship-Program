* The league was founded by the Board of Control for Cricket in India (BCCI) in 2007.


## About Dataset :

### 1. Matches Data :

* Dataset contains information about details of every ipl matches.
* It have Teams, City & Stadium in which match was played, Name of all the Umpires, Winner of toss and toss decision and also the winner.

### 2. Deliveries Data :

* Details of every Deliveries bowled in IPL.

### 3. Runs & Strikerate Data :

* Details about the Runs Scored by every Batter along with their Strike Rate and Average.

### 4. Players Data :

* Details of the Player including their DOB, Batting Hand and the Country they belonged.

### 5. Teams Data :

* Name of all the Teams that have played in the history of IPL.

### 6. Teamwise Home and Away Data :

* How many Matches does a Team Won while playing in Home Stadium and away and their percentage win-loss.

```python
def name_changer(data):
    if data == 'Delhi Daredevils':
        data = 'Delhi Capitals'
    elif data == 'Deccan Chargers':
        data = 'Sunrisers Hyderabad'
    elif data == 'Pune Warriors' or data == 'Rising Pune Supergiant':
        data = 'Rising Pune Supergiants'
    return data    
```

```python
sports['team1'] = sports['team1'].apply(lambda x:name_changer(x))
sports['team2'] = sports['team2'].apply(lambda x:name_changer(x))
sports['toss_winner'] = sports['toss_winner'].apply(lambda x:name_changer(x))
sports['winner'] = sports['winner'].apply(lambda x:name_changer(x))
```

```python
balls_bowled = pd.DataFrame(merge.groupby('bowler')['ball'].count())
wickets_taken = pd.DataFrame(merge[merge['dismissal_kind'] != 'no dismissal'].groupby('bowler')['dismissal_kind'].count())
seasons_played = pd.DataFrame(merge.groupby('bowler')['Season'].nunique())
bowler_strike_rate = pd.DataFrame({'balls':balls_bowled['ball'],'wickets':wickets_taken['dismissal_kind'],
                          'Season':seasons_played['Season']})
bowler_strike_rate.reset_index(inplace = True)

bowler_strike_rate['strike_rate'] = bowler_strike_rate['balls']/bowler_strike_rate['wickets']
def highlight_cols(s):
    color = 'skyblue'
    return 'background-color: %s' % color


best_bowling_strike_rate = bowler_strike_rate[bowler_strike_rate['wickets'] > 50].sort_values(by = 'strike_rate', ascending = True)
best_bowling_strike_rate.head(10).style.applymap(highlight_cols, subset = pd.IndexSlice[:, ['bowler', 'wickets','strike_rate']])
```

```python
plt.figure(figsize = (15,6))
sns.barplot(x = 'bowler', y = 'strike_rate', data = best_bowling_strike_rate.head(10))
plt.title("Strike rate for bowlers who have taken more than 50 wickets in IPL", fontsize = 20)
plt.xlabel("Players", size = 20)
plt.ylabel("strike_rate", size = 20)
```

* SL Malinga, DJ Bravo, A Nehra are the bowlers who took most wickets with best bowling strikerate.
